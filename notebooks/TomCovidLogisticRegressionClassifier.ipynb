# Early Detection
While using more data in the timeseries provided would likely result in a more accurate predictor, timeley diagnosis and prediction is critical in a developing crisis. Below, we only consider the data collected directly after admission to the hospital. 

```python
df = data[data["WINDOW"] == "0-2"]
print(df.shape)

df = df.dropna() #two patients have missing values. drop them from the dataset.
X = df[feature_cols]
y = df["ever_icu"]
X.head()
```

```python
#Logistic Regression
#Adapting code walkthrough from https://www.datacamp.com/community/tutorials/understanding-logistic-regression-python


X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)

print(X_train.head())
print(y_train.head())

logisticRegression = LogisticRegression(solver = 'liblinear', max_iter=1000)
logisticRegression.fit(X_train,y_train)
y_pred = logisticRegression.predict(X_test)
proba = logisticRegression.predict_proba(X_test)
y_pred2 = np.ndarray((len(proba)))
#we can tune this value to determine the threshold at which something is classified as group 0 or group 1. If we make it higher, our precision will decrease but our recall will increase.
#Would be worth in increse the threshold if we want to err on the side of predicting that someone needs ICU care, even if they don't actually. 
for i in range (0,len(proba)): 
  if(proba[i][0] > 0.7): 
    y_pred2[i] = 0
  else:
    y_pred2[i] = 1

print(y_pred)
print(y_pred2)
```

```python
#Evaluate the Logistic Regression
from sklearn import metrics
import seaborn as sns

cnf_matrix = metrics.confusion_matrix(y_test,y_pred2)
cnf_matrix

class_names=["Need ICU","Do Not Need ICU"]
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')

print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))
print("Precision:",metrics.precision_score(y_test, y_pred2))
print("Recall:",metrics.recall_score(y_test, y_pred2))

```

```python
y_pred_proba = logisticRegression.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
