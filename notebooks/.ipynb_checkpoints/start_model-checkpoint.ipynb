<h1> Modeling Baseline </h1>

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../data/quay_data.csv')
```

```python
data = data.drop(columns=[
                         'AGE_PERCENTIL',
                         'GENDER',
                         'Unnamed: 0'])
data.head()
```

<h2> For First Time Slot </h2>

```python
first = data[data['WINDOW'] == '0-2']
```

```python
### First Features

data.columns

first = first[['LACTATE_MEAN',
'RESPIRATORY_RATE_MEAN',
'CREATININ_MEAN',
'PCR_MEAN',
'AGE_ABOVE65',
'CALCIUM_MEAN',
'NEUTROPHILES_MEAN',
'HEMOGLOBIN_MEAN',
'PLATELETS_MEAN',
'HEMATOCRITE_MEAN',
'OXYGEN_SATURATION_MAX',
'LEUKOCYTES_MEAN',
'UREA_MEAN',
'OXYGEN_SATURATION_MEDIAN','ICU',
       'age_mapped', 'GENDER_0', 'GENDER_1', 'ever_icu']]
```

```python
first.head()
```

```python
len(first)  ### the bad thing about this is we actually dont have that many data points
```

```python
## For ever in ICU 

one_ever_icu = first.drop(columns=['ICU'])


## for not in icu rn

one_icu_now = first.drop(columns=['ever_icu'])
```

```python
# Ever in ICU 

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
y = one_ever_icu[['ever_icu']]
X = one_ever_icu.drop(columns=['ever_icu'])


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)



clf = LogisticRegression(random_state=0).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
clf.score(X_test, y_test)

## Gives us the accuracy scores
```

```python
### It seems that we get much higher scores as we go on

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
y = one_icu_now[['ICU']]
X = one_icu_now.drop(columns=['ICU'])


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)



clf = LogisticRegression(random_state=0).fit(X_train, y_train)
clf.predict(X_test)
clf.predict_proba(X_test)
clf.score(X_test, y_test)
```
